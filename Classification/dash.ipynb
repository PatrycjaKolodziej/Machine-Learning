## plotly - gapminder

#### Wykorzystanie widgetów interaktywnych

```python
# Wczytanie danych
gapminder_data = px.data.gapminder()

# Inicjalizacja aplikacji Dash
app = dash.Dash(__name__)

# Layout aplikacji
app.layout = html.Div([
    html.Div([
        html.H1("Analiza danych Gapminder", style={"textAlign": "center"})
    ], style={"backgroundColor": "#3c3c3c", "padding": "10px", "color": "white"}),

    html.Div([
        dcc.Dropdown(id="country_dropdown",
                     options=[{"label": country, "value": country}
                              for country in gapminder_data["country"].unique()],
                     value="Poland",
                     multi=True,
                     style={"margin": "20px"}),

        html.Div([
            html.Div([
                dcc.Graph(id="gdp_per_capita"),
            ], className="six columns"),

            html.Div([
                dcc.Graph(id="life_expectancy"),
            ], className="six columns"),
        ], className="row"),

        html.Div([
            html.Div([
                dcc.Graph(id="population"),
            ], className="six columns"),

            html.Div([
                dcc.Graph(id="gdp"),
            ], className="six columns"),
        ], className="row"),
    ], style={"backgroundColor": "#f0f0f0", "padding": "20px"})
])

# Callbacks
@app.callback(
    Output("gdp_per_capita", "figure"),
    [Input("country_dropdown", "value")])
def update_gdp_per_capita(countries):
    data = gapminder_data[gapminder_data["country"].isin(countries)]
    fig = px.line(data, x="year", y="gdpPercap", color="country")
    return fig


@app.callback(
    Output("life_expectancy", "figure"),
    [Input("country_dropdown", "value")])
def update_life_expectancy(countries):
    data = gapminder_data[gapminder_data["country"].isin(countries)]
    fig = px.line(data, x="year", y="lifeExp", color="country")
    return fig


@app.callback(
    Output("population", "figure"),
    [Input("country_dropdown", "value")])
def update_population(countries):
    data = gapminder_data[gapminder_data["country"].isin(countries)]
    fig = px.line(data, x="year", y="pop", color="country")
    return fig


@app.callback(
    Output("gdp", "figure"),
    [Input("country_dropdown", "value")])
def update_gdp(countries):
    data = gapminder_data[gapminder_data["country"].isin(countries)]
    data["gdp"] = data["gdpPercap"] * data["pop"]
    fig = px.line(data, x="year", y="gdp", color="country")
    return fig

if __name__ == "__main__":
    app.run_server(debug=False)
```

#### Parametry stylu w HTML

Parametry stylu w HTML są oparte na języku CSS (Cascading Style Sheets). Istnieje wiele atrybutów stylu, które można dostosować. Oto kilka przykładów:

* color: określa kolor tekstu w elemencie.
* backgroundColor: określa kolor tła elementu.
* border: dodaje ramkę wokół elementu, np. border: 2px solid black.
* borderRadius: zaokrągla narożniki ramki, np. borderRadius: 5px.
* fontSize: określa wielkość czcionki tekstu, np. fontSize: 18px.
* fontWeight: określa grubość czcionki, np. fontWeight: bold.
* fontStyle: określa styl czcionki, np. fontStyle: italic.
* textDecoration: określa dekorację tekstu, np. textDecoration: underline.
* textAlign: określa wyrównanie tekstu, np. textAlign: center.
* width i height: określa szerokość i wysokość elementu, np. width: 200px, height: 100px.
* margin: określa zewnętrzne odstępy między elementami, np. margin: 10px.
* padding: określa wewnętrzne odstępy między zawartością elementu a jego granicą, np. padding: 20px.
* display: określa sposób wyświetlania elementu, np. display: flex, display: inline-block, display: grid itp.
* flex: określa sposób rozkładu elementów wewnątrz kontenera flexbox, np. flex: 1, flex-direction: row, justify-content: center, align-items: center itp.
* position: określa sposób pozycjonowania elementu, np. position: absolute, position: fixed, position: relative.
* top, bottom, left, right: określa pozycję elementu względem najbliższego elementu nadrzędnego z ustawioną pozycją inną niż statyczna, np. top: 10px, left: 20px.
